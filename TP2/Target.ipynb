# Cálculo de labels, Primera ventana, Sc(d)

```python
installs = pd.read_csv('data/installs.csv.gzip', low_memory=False, compression='gzip', parse_dates = ['created'], index_col=['created'])
installs = installs.sort_index()
installs.head()
```

```python
label = installs['2019-04-21':'2019-04-23'].reset_index().groupby('ref_hash').agg({'created' : 'min'})
```

```python
label['created'] = label['created'] - pd.Timestamp('2019-04-21 00:00:00')
label['created'] = label['created'].transform(lambda x: x.total_seconds())
```
